```python
from transformers import pipeline
from transformers import BarkModel, BarkProcessor, BarkPreTrainedModel
from transformers import BarkModel, BarkProcessor, BarkPreTrainedModel

from IPython.display import Audio
from scipy.io import wavfile
```

```python
Processor = BarkProcessor.from_pretrained('suno/bark')
Model = BarkModel.from_pretrained('suno/bark')
```

```python
[k for k in Processor.speaker_embeddings.keys() if 'fr_' in k]
```

```python
inputs = Processor("Je travaille a le maison et je mange une croissant", voice_preset='fr_speaker_1')
```

```python
speech_output = Model.generate(**inputs).cpu().numpy()
```

```python
Audio(speech_output, rate=24000)
```

```python
wavfile.write("../artifacts/test.wav", rate=24000, data=speech_output.squeeze())
```
